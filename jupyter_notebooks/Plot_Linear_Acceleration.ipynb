# Linear Acceleration in the X-direction

Plot the x-direction linear acceleration from the example file: `examples/pydatabot_save_laccl_x_dir_data_to_file.py`

The data is written to:

`examples/data/linear_x_accl_data.txt`

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_json("./data/linear_x_accl_data.txt", lines=True)
```

```python
df.shape
```

```python
df.head()
```

```python
df.plot(x='time', y='linear_acceleration_x')
```

## Plot Linear X and Y Acceleration

```python
df = pd.read_json("./data/linear_x_y_accl_data.txt", lines=True)
```

```python
sns.lineplot(x='time', y='linear_acceleration_x', data=df)
gfg = sns.lineplot(x='time', y='linear_acceleration_y', data=df)

gfg.set(xlabel ="Time", ylabel = "Accl", title ='X & Y linear acceleration')
```
